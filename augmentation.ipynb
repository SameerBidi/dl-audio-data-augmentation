```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import IPython.display as ipd
```

```python
def load_audio_file(file_path, sr):
    return librosa.core.load(file_path, sr=sr)[0]

def plot_audio_data(data):
    plt.figure(figsize=(14, 8))
    plt.title("Wave")
    plt.ylabel("Amplitude")
    plt.plot(np.linspace(0, 1, len(data)), data)
    plt.show()
```

```python
sample_rate = 16000

data = load_audio_file("audios/yes.wav", sample_rate)
plot_audio_data(data)

ipd.Audio(data, rate=sample_rate)
```

#### Adding White Noise

```python
wn = np.random.randn(len(data)) * 0.005
data_wn = data + wn
plot_audio_data(data_wn)
ipd.Audio(data_wn, rate=sample_rate)
```

#### Wave Shifting

```python
data_roll = np.roll(data, int(sample_rate / 2))
plot_audio_data(data_roll)
ipd.Audio(data_roll, rate=sample_rate)
```

```python
data_stretch = librosa.effects.time_stretch(data, rate=0.5)
plot_audio_data(data_stretch)
ipd.Audio(data_stretch, rate=16000)
```

### Time Stretch

```python
data_stretch = librosa.effects.time_stretch(data, rate=2)
plot_audio_data(data_stretch)
ipd.Audio(data_stretch, rate=16000)
```
